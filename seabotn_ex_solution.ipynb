```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
tips_df = pd.read_csv(r"tips.csv")
tips_copy_df = tips_df.copy()

print('Scatter Plot')
plt.figure(figsize=(8, 6))
sns.scatterplot(data=tips_copy_df, x='total_bill', y='tip', hue="sex", palette='Spectral')
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.title('Correlation Between total bill and tip')
plt.tight_layout()
plt.legend(loc=(1.1, 0))
plt.show()
```

```python
print('Bar Chart')
plt.figure(figsize=(12, 4))
sns.barplot(data=tips_copy_df, x='smoker', y='total_bill')
plt.title('Total bill of smoker vs no smoker')
plt.show()
```

```python
print('Count Plot')
plt.figure(figsize=(8,2)) 
sns.countplot(data=tips_copy_df, x='day')
plt.title('Total bill of smoker vs no smoker')
plt.show()
```

```python
print('Histogram')
plt.figure(figsize=(6,2)) 
sns.displot(data=tips_copy_df, x='total_bill', bins=20,  kde=True)
```

```python
sns.boxplot(data=tips_copy_df, x='day', y='total_bill', hue='sex')
plt.title('Box Plot')
plt.xlabel('Day')
plt.show()
```
